**Feature Engineering**

* Feature engineering is the process of creating new features or modifying existing ones in a dataset to make it easier for machine learning models to understand and extract patterns from the data.

* It's like preparing the ingredients before cooking - you're transforming or creating variables in a way that helps the model to learn and make better predictions.


* Here we will create percentage depreciation of the laptop and create a feature indicating whether the laptop has a high or low selling price compared to its actual price.

```python
res["Depreciation (%)"] = res['Actual Price (Rs)']- res['Old Laptop Selling Price (Rs)'] / res['Actual Price (Rs)'] * 100
res["Price Status"] = res['Old Laptop Selling Price (Rs)'] > res['Actual Price (Rs)']
res["Price Status"] = res ["Price Status"].map({True:"High",False:"Low"})
print("\nDataset after Feature Engineering:")
display(res)
```

**Exaplnation:**

**Depreciation Percentage Calculation:**

It calculates the depreciation percentage of old laptops. The depreciation percentage is a measure of how much the selling price has decreased compared to the original (actual) price. The result is stored in a new column called 'Depreciation (%)'.

**Price Status Calculation:**

It determines whether the selling price of each old laptop is considered high or low compared to the actual price. The result is stored in a new column called 'Price Status'. If the selling price is higher than the actual price, the 'Price Status' is set to 'High'; otherwise, it is set to 'Low'.

**Printing the Result:**

The code then prints the DataFrame after these calculations, displaying the original data along with the newly added 'Depreciation (%)' and 'Price Status' columns.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Assuming 'res' is your DataFrame
# Initialize LabelEncoder
label_encoded = LabelEncoder()

# Encode categorical variables
res['Brand'] = label_encoded.fit_transform(res['Brand'])
res['Processor'] = label_encoded.fit_transform(res['Processor'])
res['Physical Condition'] = label_encoded.fit_transform(res['Physical Condition'])
res['Operating System'] = label_encoded.fit_transform(res['Operating System'])

# Extract numeric values from 'Storage (GB)' column and convert to float
res['Storage (GB)'] = res['Storage (GB)'].astype(float)

Features = ['RAM (GB)', 'Storage (GB)', 'Year of Release', 'Age (Years)', 
            'Screen Damage', 'Charger Availability', 'Keyword Working', 
            'Battery Health (%)']
Targets = ['Old Laptop Selling Price (Rs)']

x = res[Features]
y = res[Targets]

model = LinearRegression()
model.fit(x, y)

RM = int(input("Enter the RAM (GB): "))
ST = int(input("Enter the Storage (GB): "))
YR = int(input("Enter the Year of Release: "))
AY = int(input("Enter the Age (Years): "))
SD = int(input("Enter the Screen Damage (0 for No, 1 for Yes): "))
CA = int(input("Enter the Charger Availability (0 for No, 1 for Yes): "))
KW = int(input("Enter the Keyword Working (0 for No, 1 for Yes): "))
BH = int(input("Enter the Battery Health (%): "))

user_input = [[RM, ST, YR, AY, SD, CA, KW, BH]]

if __name__ == "__main__":
    df = pd.DataFrame(user_input, columns=Features)
    prediction = model.predict(df)
    print("Predicted Old Laptop Selling Price (Rs):", prediction[0])
```

The features (input variables) for the machine learning model are specified as ['RAM (GB)', 'Storage (GB)', 'Year of Release', 'Age (Years)', 'Screen Damage', 'Charger Availability', 'Keyword Working', 'Battery Health (%)'].
The target variable (output to be predicted) is set as 'Old Laptop Selling Price (Rs)'.

**Extract Features and Target:**

The features (X) and the target variable (y) are extracted from the DataFrame (df) using the specified column names.
